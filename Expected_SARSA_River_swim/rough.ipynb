```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
num_episodes = 10
reward = np.random.rand(num_episodes)  # Random rewards for illustration

# Calculate cumulative sum
reward_cumsum = np.cumsum(reward)

# Calculate running average and standard deviation for error bars
running_average = reward_cumsum / np.arange(1, num_episodes + 1)
std_reward = np.std(reward_cumsum) / np.sqrt(np.arange(1, num_episodes + 1))

# Plotting with error bars
plt.errorbar(np.arange(0, num_episodes), reward_cumsum, yerr=std_reward, fmt='o', capsize=5, capthick=2, elinewidth=1)
plt.title('Cumulative Reward with Error Bars')
plt.xlabel('Episode')
plt.ylabel('Cumulative Reward')
plt.grid(True)
plt.show()
```
